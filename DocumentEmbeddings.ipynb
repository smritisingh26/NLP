## WHAT IS DOCUMENT EMBEDDING?

Document Embedding is essentially an extension of the concept of word embeddings, where words are mapped into numerical vector spaces. Using word embeddings, words that have the same meaning have similar vectorial representation in space. While word embeddings focus on mapping words to vector spaces, document embedding focuses on mapping entire documents (sentences, articles or even books) into numerical vector spaces. 

The applications of document embeddings include but are not limited to:
1. Text classification
2. Sentiment analysis
3. Question answering
4. Image captioning
5. Machine translation

### 1. USING DOC2VEC FOR DOCUMENT EMBEDDINGS

Just like document embedding is an extension of word embedding, doc2vec is an extension of the algorithm word2vec. So to understand doc2vec, let's first take a look at word2vec.

word2vec aims to create word embeddings by using two algorithms- Continuous bag of words (CBOW) and the skip gram model.The goal of doc2vec is to create a vectorial representation of a document, regardless of its length. But unlike words, documents do not always come in logical structures such as words. 

Thus, doc2vec uses an algorithm called Distributed Memory version of Paragraph Vector (PV-DM). It acts as a memory that remembers what is missing from the current context — or as the topic of the paragraph. While the word vectors represent the concept of a word, the document vector intends to represent the concept of a document. 

For the implementation of doc2vec, we'll use gensim. 

```python
#importing necessary libraries 
from gensim.models.doc2vec import Doc2Vec,TaggedDocument
from os import listdir
from os.path import isfile, join
from nltk import RegexpTokenizer
from nltk.corpus import stopwords
from os import listdir
from os.path import isfile, join
```

```python
#Now, we'll use the articles that were loaded into separate text files. 
tokenizer=RegexpTokenizer(r'\w+')
stopword_set = set(stopwords.words('english'))
def clean(data):
    clean_data=[]
    for i in data:
        new_str=i.lower()
        mylist=tokenizer.tokenize(new_str)
        mylist=list(set(mylist).difference(stopword_set))
        clean_data.append(mylist)
    return clean_data
```

```python
class LabeledLineSentence(object):
    def __init__(self, doc_list, labels_list):
        self.labels_list=labels_list
        self.doc_list=doc_list
    def __iter__(self):
        for idx, doc in enumerate(self.doc_list):
              yield gensim.models.doc2vec.LabeledSentence(doc,[self.labels_list[idx]])
```

```python
data=clean(data)
```

```python
itr=LabeledLineSentence(data,docLabels)
```

```python
import gensim
mymodel=gensim.models.Doc2Vec(size=300, min_count=0, alpha=0.025, min_alpha=0.025)
mymodel.build_vocab(itr)
#training of model
for epoch in range(100):
    print('iteration'+str(epoch+1))
    mymodel.train(itr,total_examples=mymodel.corpus_count,epochs=mymodel.epochs)
    mymodel.alpha -= 0.002
    mymodel.min_alpha=mymodel.alpha
mymodel.save('doc2vec.model')
print("model saved.")
```

### 2. TOPIC MODELLING USING LDA

Topic modeling is a type of statistical modeling for discovering the abstract “topics” that occur in a collection of documents. Latent Dirichlet Allocation (LDA) is an example of an algorithm used for topic modelling and is used to classify text in a document to a particular topic. It builds a topic per document model and words per topic model, modeled as Dirichlet distributions.

While Document Embedding is not usually the main application for topic modelling,it inherently generates a document embedding space meant to model and explain word distribution in the corpus and where dimensions can be seen as latent semantic structures hidden in the data.

For the purpose of a sample implementation, we'll be using sklearn.

```python
# Create and fit the LDA model
from sklearn.decomposition import LatentDirichletAllocation as LDA
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
# Fit and transform the data 
count_vectorizer=CountVectorizer()
num_data=count_vectorizer.fit_transform(filtered_doc)
lda=LDA(n_components=20, n_jobs=-1)
#data=data.reshape(-1,1)
lda.fit(num_data)
```

```python
print("Topics found via LDA:")
def print_topics(model, count_vectorizer, n_top_words):
    words=count_vectorizer.get_feature_names()
    for topic_idx, topic in enumerate(model.components_):
        print("\nTopic #%d:" % topic_idx)
        print(" ".join([words[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
print_topics(lda, count_vectorizer,n_top_words=10)
```

This worked too! The model has picked up the key ideas of the articles and converted them into topics. What's more is, it had to convert each document into a vector before it could achieve this. Thus, LDA involves "latent" document embedding. Let's look at one more method of document embedding before we move on to building a text classifier.

```python
#Building the feature vector for the text classifier
def feature_vec(model,tagged_docs):
    sents=tagged_docs.values
    targets,regs=zip(*[(doc.tags[0], model.infer_vector(doc.words, steps=20)) for doc in sents])
    return targets,regs
```
